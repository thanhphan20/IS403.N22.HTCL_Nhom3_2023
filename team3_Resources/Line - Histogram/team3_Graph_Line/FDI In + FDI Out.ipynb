```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "/content/sample_data/data.csv"
df = pd.read_csv(path)
df = df.iloc[1:]
```

```python
df
```

```python
df['year'] = df['year'].astype('int')
df['GDP'] = df['GDP'].astype('float')
df['agriculture'] = df['agriculture'].astype('float')
df['industry'] = df['industry'].astype('float')
df['service'] = df['service'].astype('float')
df['CPI'] = df['CPI'].astype('float')
df['FDI in'] = df['FDI in'].astype('float')
df['FDI out'] = df['FDI out'].astype('float')
df['inflation rate'] = df['inflation rate'].astype('float')
df['exchange rate'] = df['exchange rate'].astype('float')
df['exchange'] = df['exchange'].astype('float')
df['export'] = df['export'].astype('float')
df['import'] = df['import'].astype('float')
df['workers'] = df['workers'].astype('float')
df['unemployer rate'] = df['unemployer rate'].astype('float')
```

```python
#Phân tích FDI In - FDI Out 2000-20023 theo năm
filtered_data_2000 = df[(df['year'] >= 2000) & (df['year'] <= 2023)]

fdi_in_out_2000 = filtered_data_2000.groupby('year')[['FDI in','FDI out']].sum().reset_index()

df_fdi_in_out_2000 = pd.DataFrame(fdi_in_out_2000)

df_fdi_in_out_2000
```

```python
# Vẽ biểu đồ cho FDI In và FDI Out
plt.figure(figsize=(10, 6))
plt.plot(df_fdi_in_out_2000['year'], df_fdi_in_out_2000['FDI in'], marker='o', label='FDI in')
plt.plot(df_fdi_in_out_2000['year'], df_fdi_in_out_2000['FDI out'], marker='o', label='FDI out')

plt.title('FDI In/FDI Out (Tỷ USD / Năm)')
plt.xlabel('Năm')
plt.ylabel('FDI In/FDI Out (Tỷ USD)')

plt.legend()
plt.grid(True)
plt.show()
```

```python
#Phân tích FDI In 2000-20023 theo năm
filtered_data_2000 = df[(df['year'] >= 2000) & (df['year'] <= 2023)]

fdi_in_2000 = filtered_data_2000.groupby('year')[['FDI in']].sum().reset_index()

df_fdi_in_2000 = pd.DataFrame(fdi_in_2000)

df_fdi_in_2000
```

```python
# Vẽ biểu đồ cho FDI In
plt.figure(figsize=(10, 6))
plt.plot(df_fdi_in_2000['year'], df_fdi_in_2000['FDI in'], marker='o', label='FDI In')
plt.title('FDI In (Billion USD / năm)')
plt.xlabel('Năm')
plt.ylabel('Số tiền đầu tư (Billion USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Phân tích FDI Out 2000-20023 theo năm
filtered_data_2000 = df[(df['year'] >= 2000) & (df['year'] <= 2023)]

fdi_out_2000 = filtered_data_2000.groupby('year')[['FDI out']].sum().reset_index()

df_fdi_out_2000 = pd.DataFrame(fdi_out_2000)

df_fdi_out_2000
```

```python
# Vẽ biểu đồ cho FDI Out
plt.figure(figsize=(10, 6))
plt.plot(df_fdi_out_2000['year'], df_fdi_out_2000['FDI out'], marker='o', label='FDI Out')
plt.title('FDI Out (Billion USD / năm)')
plt.xlabel('Năm')
plt.ylabel('Số tiền đầu tư (Billion USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Phân tích FDI In/FDI Out 2000-2023 theo tháng
filtered_data_2000 = df[(df['year'] >= 2000) & (df['year'] <= 2023)]

filtered_data_2000 = filtered_data_2000[['year','month', 'FDI in', 'FDI out']]

filtered_data_2000 = filtered_data_2000.groupby(['year', 'month']).head()

df_fdi_in_out_2000 = pd.DataFrame(filtered_data_2000)

df_fdi_in_out_2000
```

```python
plt.figure(figsize=(36, 6))  # Kích thước biểu đồ

# Vẽ đường cho giá trị FDI In
plt.plot(df_fdi_in_out_2000.index, df_fdi_in_out_2000['FDI in'], marker='o', label='FDI In', linestyle='-', color='b')

# Vẽ đường cho giá trị FDI Out
plt.plot(df_fdi_in_out_2000.index, df_fdi_in_out_2000['FDI out'], marker='o', label='FDI Out', linestyle='-', color='r')

# Đặt tên cho trục x và y
plt.xticks(df_fdi_in_out_2000.index, df_fdi_in_out_2000['year'].astype('str') + ' ' + df_fdi_in_out_2000['month'], rotation=90)
plt.xlabel('Thời gian')
plt.ylabel('Giá trị FDI In/FDI Out (Tỷ USD)')

plt.title('Biểu đồ giá trị FDI In/FDI Out từ 2000 đến 2023')
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

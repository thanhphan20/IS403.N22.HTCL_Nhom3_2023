```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df = df.iloc[1:]
df
```

```python
df['year'] = df['year'].astype('int')
df['CPI'] = df['CPI'].astype('float')
```

```python
#Phân tích CPI 2000-2023 theo tháng
filtered_data_2000 = df[(df['year'] >= 2000) & (df['year'] <= 2023)]

filtered_data_2000 = filtered_data_2000[['year', 'month', 'CPI']]

filtered_data_2000 = filtered_data_2000.groupby(['year', 'month']).head()

df_cpi_2000 = pd.DataFrame(filtered_data_2000)

df_cpi_2000
```

```python
plt.figure(figsize=(60, 10))  # Kích thước biểu đồ

plt.plot(df_cpi_2000.index, df_cpi_2000['CPI'], marker='o', label='CPI', linestyle='-', color='b')
plt.xticks(df_cpi_2000.index, df_cpi_2000['year'].astype('str') + ' ' + df_cpi_2000['month'], rotation=90)
plt.xlabel('Thời gian')
plt.ylabel('Giá trị CPI %')

plt.title('Biểu đồ giá trị CPI từ 2000 đến 2023')
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

```python
#Phân tích CPI từ năm 2000-2010
filtered_data_2000 = df[(df['year'] >= 2000) & (df['year'] <= 2010)]

df_cpi_2000 = filtered_data_2000.groupby('year')[['CPI']].sum().reset_index()

df_cpi_2000 = pd.DataFrame(df_cpi_2000)

df_cpi_2000
```

```python
# Vẽ biểu đồ cho CPI
plt.figure(figsize=(10, 6))
plt.plot(df_cpi_2000['year'], df_cpi_2000['CPI'], marker='o', label='CPI')

plt.title('Chỉ số CPI từ 2000-2010 (%/ Năm)')
plt.xlabel('Năm')
plt.ylabel('Chỉ số CPI (%)')

plt.legend()
plt.grid(True)
plt.show()
```
